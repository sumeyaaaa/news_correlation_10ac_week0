```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.parse import urlparse

# Function to extract the main domain from a URL
def extract_main_domain(url):
    try:
        parsed_url = urlparse(url)
        domain = parsed_url.netloc
        # Remove the 'www.' prefix if it exists
        if domain.startswith('www.'):
            domain = domain[4:]
        # Extract the two parts connected by a dot
        domain_parts = domain.split('.')
        if len(domain_parts) > 2:
            domain = '.'.join(domain_parts[-2:])
        return domain
    except Exception as e:
        return None

# Define the column name for URLs
url_column_name = 'url'

# Apply the function to extract the main domain from the URL column
merged_df['Main_Domain'] = merged_df[url_column_name].apply(extract_main_domain)

# Define a mapping for sentiment values
sentiment_mapping = {'Positive': 1, 'Neutral': 0, 'Negative': -1}

# Convert title_sentiment to numeric using the mapping
merged_df['title_sentiment'] = merged_df['title_sentiment'].map(sentiment_mapping)

# Handle missing values in sentiment column
merged_df['title_sentiment'].fillna(0, inplace=True)

# Group by 'Main_Domain' and count the number of articles
report_counts = merged_df['Main_Domain'].value_counts().reset_index()
report_counts.columns = ['Main_Domain', 'Total_Reports']

# Compute average sentiment for each domain
average_sentiment = merged_df.groupby('Main_Domain')['title_sentiment'].mean().reset_index()
average_sentiment.columns = ['Main_Domain', 'Average_Sentiment']

# Prepare db_traffic DataFrame
db_traffic['Domain'] = db_traffic['Domain'].apply(lambda x: x.replace('www.', '') if x.startswith('www.') else x)

# Merge with db_traffic to get GlobalRank
merged_with_traffic = pd.merge(report_counts, average_sentiment, on='Main_Domain', how='left')
merged_with_traffic = pd.merge(merged_with_traffic, db_traffic[['Domain', 'GlobalRank']], 
                               left_on='Main_Domain', right_on='Domain', 
                               how='left')

# Drop unnecessary columns
merged_with_traffic = merged_with_traffic.drop(columns=['Domain'])

# Set up the visual aesthetics
sns.set(style="whitegrid")

# Create the scatter plot
fig, ax = plt.subplots(figsize=(14, 8))
scatter_plot = sns.scatterplot(
    x='Total_Reports', 
    y='GlobalRank', 
    hue='Average_Sentiment', 
    palette='coolwarm',  # Color palette for sentiment
    data=merged_with_traffic,
    edgecolor=None,
    ax=ax,
    legend='full'
)

# Add titles and labels
ax.set_title('Scatter Plot of Reports vs Global Ranking by Website Sentiment', fontsize=16, fontweight='bold')
ax.set_xlabel('Total Number of Reports', fontsize=14)
ax.set_ylabel('Global Ranking', fontsize=14)
ax.set_xscale('log')  # Use logarithmic scale if there's a wide range of report counts
ax.set_yscale('log')  # Use logarithmic scale if global rankings vary widely

# Add a color bar
norm = plt.Normalize(vmin=-1, vmax=1)
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='vertical')
cbar.set_label('Average Sentiment', fontsize=14)

# Add website names below the graph (first 10)
for i, row in merged_with_traffic.head(30).iterrows():
    ax.text(row['Total_Reports'], row['GlobalRank'], row['Main_Domain'], fontsize=9, ha='right')

# Calculate sentiment counts and averages
sentiment_counts = merged_df['title_sentiment'].value_counts()
positive_count = sentiment_counts.get(1, 0)
negative_count = sentiment_counts.get(-1, 0)
neutral_count = sentiment_counts.get(0, 0)

average_positive = average_sentiment.loc[average_sentiment['Average_Sentiment'] == 1, 'Average_Sentiment'].mean()
average_negative = average_sentiment.loc[average_sentiment['Average_Sentiment'] == -1, 'Average_Sentiment'].mean()
average_neutral = average_sentiment.loc[average_sentiment['Average_Sentiment'] == 0, 'Average_Sentiment'].mean()

print(f"Positive Sentiments Count: {positive_count}")
print(f"Negative Sentiments Count: {negative_count}")
print(f"Neutral Sentiments Count: {neutral_count}")
print(f"Average Positive Sentiment: {average_positive}")
print(f"Average Negative Sentiment: {average_negative}")
print(f"Average Neutral Sentiment: {average_neutral}")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report



# Extract the 'article' column for analysis
headlines =  merged_df['article'].fillna('')  # Fill NaN with empty strings

# Step 2: Define Keywords for Each Category
categories = {
    'Breaking News': ['breaking', 'alert', 'urgent'],
    'Politics': ['election', 'policy', 'congress', 'senate'],
    'World News': ['international', 'global', 'UN', 'war', 'diplomacy'],
    'Business/Finance': ['market', 'stocks', 'finance', 'economy'],
    'Technology': ['tech', 'software', 'hardware', 'AI', 'cyber'],
    'Science': ['research', 'study', 'scientific', 'discovery'],
    'Health': ['health', 'medicine', 'wellness', 'disease'],
    'Entertainment': ['celebrity', 'movie', 'TV', 'music'],
    'Sports': ['match', 'game', 'tournament', 'player'],
    'Environment': ['climate', 'ecology', 'environment'],
    'Crime': ['crime', 'arrest', 'police', 'court','kill','kidnapped'],
    'Education': ['school', 'university', 'education', 'study'],
    'Weather': ['weather', 'storm', 'hurricane', 'forecast'],
    'Other': []  # Default category for unmatched articles
}

# Step 3: Function to Tag Articles Based on Keywords
def assign_category(article):
    for category, keywords in categories.items():
        if any(keyword in article.lower() for keyword in keywords):
            return category
    return 'Other'  # Default category

# Apply the function to assign categories
merged_df['predicted_category'] =  merged_df['article'].apply(assign_category)

# Step 4: TF-IDF Vectorization and Apply LSA
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')
lsa = TruncatedSVD(n_components=100)

# Create a pipeline for SVM classification after LSA
classifier = SVC(kernel='linear')
model = make_pipeline(tfidf, lsa, classifier)

# Prepare data for training and testing
labels = merged_df['predicted_category'].values
X_train, X_test, y_train, y_test = train_test_split(headlines, labels, test_size=0.2, random_state=42)

# Train the classifier
model.fit(X_train, y_train)

# Predict and evaluate the classifier
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Step 1: Extract Title and Article Texts and Use a Smaller Sample
# Use only a sample of 10,000 rows to avoid memory issues
sampled_df = merged_df.sample(n=10000, random_state=42)
titles = sampled_df['title'].fillna('')
articles = sampled_df['article'].fillna('')

# Step 2: TF-IDF Vectorization with Reduced Number of Features
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=500)  # Reduce to 500 features

# Combine the titles and articles for joint fitting
combined_texts = titles.tolist() + articles.tolist()

# Fit and transform the combined texts
tfidf_matrix = tfidf_vectorizer.fit_transform(combined_texts)

# Separate the transformed vectors back into titles and articles
tfidf_titles = tfidf_matrix[:len(titles)]
tfidf_articles = tfidf_matrix[len(titles):]

# Step 3: Compute Cosine Similarity
# Calculate cosine similarity between each title and its corresponding article
similarity_scores = cosine_similarity(tfidf_titles, tfidf_articles)

# Extract the diagonal similarity scores
similarity_scores_diagonal = [similarity_scores[i, i] for i in range(len(titles))]

# Step 4: Add the Similarity Scores to the DataFrame
sampled_df['similarity_score'] = similarity_scores_diagonal

# Print the DataFrame with the new similarity scores
print(sampled_df[['title', 'article', 'similarity_score']].head())

# Step 5: Visualize the Similarity Scores
plt.figure(figsize=(10, 6))
plt.hist(sampled_df['similarity_score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Similarity Scores between Titles and Articles')
plt.xlabel('Similarity Score')
plt.ylabel('Frequency')
plt.show()

# Step 6: Analyze the Average and Median Similarity Scores
average_similarity = sampled_df['similarity_score'].mean()
median_similarity = sampled_df['similarity_score'].median()

print(f"Average Similarity Score: {average_similarity:.2f}")
print(f"Median Similarity Score: {median_similarity:.2f}")
```

### Description of the Graph

**X-Axis (Similarity Score):**  
A score closer to 1 indicates high similarity, meaning the keywords or content in the title and the article body are very similar.  
A score closer to 0 suggests low similarity, implying the content in the title and the article body differs significantly.

**Y-Axis (Frequency):**  
The y-axis shows the frequency of these similarity scores, or how often each range of scores occurs in the dataset.

**Histogram Bars:**  
Each bar represents the number of articles within a certain range of similarity scores. Taller bars indicate a higher frequency of articles having a similarity score within that specific range.

As the bars are spread more evenly or skewed towards the lower end (closer to 0), this suggests that many titles do not closely match the content of their articles, potentially indicating clickbait or misleading headlines.

### Additional Explanation

- **Average Similarity Score: 0.30**  
  This indicates that, on average, there is a low level of similarity between the keywords in the titles and the content of the articles across the dataset. A score of 0.30 suggests that, overall, the titles only moderately reflect the content in the articles.

- **Median Similarity Score: 0.29**  
  The median similarity score being close to the average (0.29) reinforces that the majority of articles have a low to moderate level of similarity between their titles and content. This further supports the idea that many headlines may not accurately represent the article's body content.
