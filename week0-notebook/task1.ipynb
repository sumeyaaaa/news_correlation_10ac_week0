```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.parse import urlparse

# Function to extract the main domain from a URL
def extract_main_domain(url):
    try:
        parsed_url = urlparse(url)
        domain = parsed_url.netloc
        # Remove the 'www.' prefix if it exists
        if domain.startswith('www.'):
            domain = domain[4:]
        # Extract the two parts connected by a dot
        domain_parts = domain.split('.')
        if len(domain_parts) > 2:
            domain = '.'.join(domain_parts[-2:])
        return domain
    except Exception as e:
        return None

# Define the column name for URLs
url_column_name = 'url'

# Apply the function to extract the main domain from the URL column
merged_df['Main_Domain'] = merged_df[url_column_name].apply(extract_main_domain)

# Define a mapping for sentiment values
sentiment_mapping = {'Positive': 1, 'Neutral': 0, 'Negative': -1}

# Convert title_sentiment to numeric using the mapping
merged_df['title_sentiment'] = merged_df['title_sentiment'].map(sentiment_mapping)

# Handle missing values in sentiment column
merged_df['title_sentiment'].fillna(0, inplace=True)

# Group by 'Main_Domain' and count the number of articles
report_counts = merged_df['Main_Domain'].value_counts().reset_index()
report_counts.columns = ['Main_Domain', 'Total_Reports']

# Compute average sentiment for each domain
average_sentiment = merged_df.groupby('Main_Domain')['title_sentiment'].mean().reset_index()
average_sentiment.columns = ['Main_Domain', 'Average_Sentiment']

# Prepare db_traffic DataFrame
db_traffic['Domain'] = db_traffic['Domain'].apply(lambda x: x.replace('www.', '') if x.startswith('www.') else x)

# Merge with db_traffic to get GlobalRank
merged_with_traffic = pd.merge(report_counts, average_sentiment, on='Main_Domain', how='left')
merged_with_traffic = pd.merge(merged_with_traffic, db_traffic[['Domain', 'GlobalRank']], 
                               left_on='Main_Domain', right_on='Domain', 
                               how='left')

# Drop unnecessary columns
merged_with_traffic = merged_with_traffic.drop(columns=['Domain'])

# Set up the visual aesthetics
sns.set(style="whitegrid")

# Create the scatter plot
fig, ax = plt.subplots(figsize=(14, 8))
scatter_plot = sns.scatterplot(
    x='Total_Reports', 
    y='GlobalRank', 
    hue='Average_Sentiment', 
    palette='coolwarm',  # Color palette for sentiment
    data=merged_with_traffic,
    edgecolor=None,
    ax=ax,
    legend='full'
)

# Add titles and labels
ax.set_title('Scatter Plot of Reports vs Global Ranking by Website Sentiment', fontsize=16, fontweight='bold')
ax.set_xlabel('Total Number of Reports', fontsize=14)
ax.set_ylabel('Global Ranking', fontsize=14)
ax.set_xscale('log')  # Use logarithmic scale if there's a wide range of report counts
ax.set_yscale('log')  # Use logarithmic scale if global rankings vary widely

# Add a color bar
norm = plt.Normalize(vmin=-1, vmax=1)
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='vertical')
cbar.set_label('Average Sentiment', fontsize=14)

# Add website names below the graph (first 10)
for i, row in merged_with_traffic.head(30).iterrows():
    ax.text(row['Total_Reports'], row['GlobalRank'], row['Main_Domain'], fontsize=9, ha='right')

# Calculate sentiment counts and averages
sentiment_counts = merged_df['title_sentiment'].value_counts()
positive_count = sentiment_counts.get(1, 0)
negative_count = sentiment_counts.get(-1, 0)
neutral_count = sentiment_counts.get(0, 0)

average_positive = average_sentiment.loc[average_sentiment['Average_Sentiment'] == 1, 'Average_Sentiment'].mean()
average_negative = average_sentiment.loc[average_sentiment['Average_Sentiment'] == -1, 'Average_Sentiment'].mean()
average_neutral = average_sentiment.loc[average_sentiment['Average_Sentiment'] == 0, 'Average_Sentiment'].mean()

print(f"Positive Sentiments Count: {positive_count}")
print(f"Negative Sentiments Count: {negative_count}")
print(f"Neutral Sentiments Count: {neutral_count}")
print(f"Average Positive Sentiment: {average_positive}")
print(f"Average Negative Sentiment: {average_negative}")
print(f"Average Neutral Sentiment: {average_neutral}")

# Show the plot
plt.tight_layout()
plt.show()
```
