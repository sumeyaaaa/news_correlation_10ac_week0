```python
import psycopg2
from psycopg2 import OperationalError

# Function to connect to the database
def get_connection():
    return psycopg2.connect(
        dbname="event_model",
        user="sumeya",
        password="your_password",
        host="localhost",
        port="5432"
    )

# Function to create a table in the database
def create_table():
    try:
        conn = get_connection()
        with conn.cursor() as cursor:
            # Define the schema with schema name
            create_table_query = '''
            CREATE TABLE IF NOT EXISTS public.event_models (
                id SERIAL PRIMARY KEY,
                text TEXT NOT NULL,
                category VARCHAR(255) NOT NULL,
                input_ids TEXT,
                attention_masks TEXT
            );
            '''
            cursor.execute(create_table_query)
            conn.commit()  # Commit the transaction
            print("Table created successfully")
    except OperationalError as e:
        print(f"OperationalError: {e}")
    finally:
        if conn:
            conn.close()

# Run the function to create the table
create_table()
```

**Y-Axis (Frequency):**

Each bar represents the number of articles within a certain range of similarity scores. Taller bars indicate a higher frequency of articles having a similarity score within that specific range.

```python
# Function to connect to the database
def get_connection():
    return psycopg2.connect(
        dbname="event_model",
        user="sumeya",
        password="your_password",
        host="localhost",
        port="5432"
    )

# Function to insert data into the database
from psycopg2 import OperationalError

def insert_data(titles, descriptions, urls, published_ats, contents, categories, articles, title_sentiments):
    try:
        conn = get_connection()
        with conn.cursor() as cursor:
            insert_query = '''
            INSERT INTO event_models (title, description, url, published_at, content, category, article, title_sentiment)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
            '''
            for title, description, url, published_at, content, category, article, title_sentiment in zip(
                    titles, descriptions, urls, published_ats, contents, categories, articles, title_sentiments):
                cursor.execute(insert_query, (title, description, url, published_at, content, category, article, title_sentiment))
            conn.commit()
            print("Data inserted successfully")
    except OperationalError as e:
        print(f"OperationalError: {e}")
    finally:
        if conn:
            conn.close()
```
